### CODE FINAL ###

```python
villes_incontournables = ["Mont Saint Michel",
"St Malo",
"Bayeux",
"Le Havre",
"Rouen",
"Paris",
"Amiens",
"Lille",
"Strasbourg",
"Chateau du Haut Koenigsbourg",
"Colmar",
"Eguisheim",
"Besancon",
"Dijon",
"Annecy",
"Grenoble",
"Lyon",
"Gorges du Verdon",
"Bormes les Mimosas",
"Cassis",
"Marseille",
"Aix en Provence",
"Avignon",
"Uzes",
"Nimes",
"Aigues Mortes",
"Saintes Maries de la mer",
"Collioure",
"Carcassonne",
"Ariege",
"Toulouse",
"Montauban",
"Biarritz",
"Bayonne",
"La Rochelle"]

load_dotenv()
api_key = os.getenv("api_key")

data_frames = []

print("\n" + "="*80)
print("Démarrage du traitement")
print("="*80)

for id_ville, ville in enumerate(villes_incontournables, start=1):
    print(f"Traitement de {ville}...")
    
    # ÉTAPE 1 : Récupérer les coordonnées avec Nominatim
    url_geo = "https://nominatim.openstreetmap.org/search"
    params_geo = {
        "q": ville,
        "format": "json",
        "addressdetails": 1,
        "limit": 1,
        "email": "julien.charlier31@gmail.com"  
    }
    headers = {"User-Agent": "MonApp/1.0"}
    
    response_geo = requests.get(url_geo, params=params_geo, headers=headers)
    
    if response_geo.status_code == 200:
        data_geo = response_geo.json()
        
        if data_geo:
            result = data_geo[0]
            la = float(result['lat'])
            lo = float(result['lon'])
            country = result.get('address', {}).get('country', 'Unknown')
            
            # ÉTAPE 2 : Récupérer la météo avec One Call API 3.0 (prévisions 7 jours)
            url_weather = "https://api.openweathermap.org/data/3.0/onecall"
            params_weather = {
                "lat": la,
                "lon": lo,
                "appid": api_key,
                "units": "metric",
                "exclude": "minutely,hourly"
            }
            
            response_weather = requests.get(url_weather, params=params_weather)
            
            if response_weather.status_code == 200:
                data_meteo = response_weather.json()
                
                # ÉTAPE 3 : Calculer le score sur les 7 prochains jours
                daily_data = data_meteo.get("daily", [])[:7]
                
                if daily_data:
                    # Moyennes sur 7 jours
                    avg_temp = sum(d["temp"]["day"] for d in daily_data) / len(daily_data)
                    avg_humidity = sum(d["humidity"] for d in daily_data) / len(daily_data)
                    total_rain = sum(d.get("rain", 0) for d in daily_data)
                    avg_pop = sum(d.get("pop", 0) for d in daily_data) / len(daily_data)
                    
                    rain_score = max(0, 100 - (total_rain * 10))
                    temp_score = max(0, 100 - abs(avg_temp - 20) * 5)
                    humidity_score = max(0, 100 - (avg_humidity - 40) * 2)
                    
                    beauty_score = (rain_score + temp_score + humidity_score) / 3
                    
                    df = pd.DataFrame({
                        "id": [id_ville],
                        "city": [ville],
                        "latitude": [la],
                        "longitude": [lo],
                        "country": [country],
                        "avg_temp_7j(°C)": [round(avg_temp, 2)],
                        "avg_humidity_7j(%)": [round(avg_humidity, 2)],
                        "total_rain_7j(mm)": [round(total_rain, 2)],
                        "avg_pop_7j(%)": [round(avg_pop * 100, 2)],
                        "beauty_score": [round(beauty_score, 2)]
                    })
                    
                    data_frames.append(df)
                    print(f"✓ {ville} - Score: {beauty_score:.2f}")
                else:
                    print(f"Pas de données daily pour {ville}")
            elif response_weather.status_code == 401:
                print(f"Erreur authentification ou clé API invalide pour {ville}")
            elif response_weather.status_code == 429:
                print(f"Rate limit atteint pour {ville}")
            else:
                print(f"Erreur OpenWeather pour {ville}: {response_weather.status_code}")
        else:
            print(f"Aucun résultat trouvé pour {ville}")
    else:
        print(f"Erreur Nominatim pour {ville}: {response_geo.status_code}")
    
    time.sleep(2)

# ÉTAPE 4 : Combiner tous les DataFrames et sauvegarder en CSV
if data_frames:
    df_final = pd.concat(data_frames, ignore_index=True)
    df_final = df_final.sort_values("beauty_score", ascending=False).reset_index(drop=True)
    
    df_final.to_csv("villes_meteo.csv", index=False, encoding='utf-8')
    
    print("\n" + "="*80)
    print("✓ DONNÉES SAUVEGARDÉES DANS 'villes_meteo.csv'")
    print("="*80)
    print(f"Total de villes traitées: {len(df_final)}\n")
       
    print("\n" + "="*80)
    print("TOP 10 - MEILLEURES DESTINATIONS PAR BEAUTY SCORE")
    print("="*80)
    top10 = df_final[['id', 'city', 'avg_temp_7j(°C)', 'avg_humidity_7j(%)', 'beauty_score']].head(10)
    top10.index = range(1, 11)
    print(top10.to_string())    
else:
    print("Aucune donnée récupérée")

# ÉTAPE 5 : Data Visualization avec Plotly
# Normaliser les tailles
df_beauty = pd.read_csv("villes_meteo.csv")
temp_min = df_beauty['avg_temp_7j(°C)'].min()
temp_max = df_beauty['avg_temp_7j(°C)'].max()
df_beauty['sizes'] = 5 + (df_beauty['avg_temp_7j(°C)'] - temp_min) / (temp_max - temp_min) * 20

score_min = df_beauty['beauty_score'].min()
score_max = df_beauty['beauty_score'].max()
df_beauty['sizes_score'] = 5 + (df_beauty['beauty_score'] - score_min) / (score_max - score_min) * 20

# CARTE 1 : Température moyenne 7 jours
fig1 = px.scatter_map(
    df_beauty,
    lat='latitude',
    lon='longitude',
    hover_name='city',
    color='avg_temp_7j(°C)',
    size='sizes',
    color_continuous_scale=px.colors.cyclical.IceFire,
    size_max=30,
    zoom=4,
    map_style='carto-positron'
)
fig1.update_layout(
    title='Gradient de Température (Moyenne 7 jours)',
    height=700,
    width=650,
    margin=dict(l=0, r=0, b=0, t=50)
)

# CARTE 2 : Beauty Score
fig2 = px.scatter_map(
    df_beauty,
    lat='latitude',
    lon='longitude',
    hover_name='city',
    color='beauty_score',
    size='sizes_score',
    color_continuous_scale='Viridis',
    size_max=30,
    zoom=4,
    map_style='carto-positron'
)
fig2.update_layout(
    title='Gradient de Beauty Score (7 jours)',
    height=700,
    width=650,
    margin=dict(l=0, r=0, b=0, t=50)
)

# Afficher les deux
fig1.show()
fig2.show()
```

```python
# Normaliser les tailles
temp_min = df_beauty['avg_temp_7j(°C)'].min()
temp_max = df_beauty['avg_temp_7j(°C)'].max()
df_beauty['sizes'] = 5 + (df_beauty['avg_temp_7j(°C)'] - temp_min) / (temp_max - temp_min) * 20

score_min = df_beauty['beauty_score'].min()
score_max = df_beauty['beauty_score'].max()
df_beauty['sizes_score'] = 5 + (df_beauty['beauty_score'] - score_min) / (score_max - score_min) * 20

# Créer le texte hover avec les hôtels
def get_hotels_text(city):
    hotels = df_top3_hotels[df_top3_hotels['Ville'] == city]
    if len(hotels) == 0:
        return "Aucun hôtel trouvé"
    
    text = "<b>Top 3 Hôtels:</b><br>"
    for idx, row in hotels.iterrows():
        text += f"{idx+1}. <b>{row['Hotel']}</b> - {row['Note']}/10<br>"
    return text

df_beauty['hotels_text'] = df_beauty['city'].apply(get_hotels_text)


fig1 = px.scatter_map(
    df_beauty,
    lat='latitude',
    lon='longitude',
    hover_name='city',
    color='avg_temp_7j(°C)',
    size='sizes',
    color_continuous_scale='RdBu_r',
    size_max=30,
    zoom=4,
    map_style='carto-positron',
    hover_data={'latitude': False, 'longitude': False, 'sizes': False}
)

fig1.update_traces(
    customdata=df_beauty['hotels_text'],
    hovertemplate='<b>%{hovertext}</b><br>Temp: %{marker.color:.1f}°C<br>%{customdata}<extra></extra>'
)

fig1.update_layout(
    title='Gradient de Température + Top 3 Hôtels',
    height=700,
    width=900,
    margin=dict(l=0, r=0, b=0, t=50),
    coloraxis=dict(
        cmin=-15,
        cmax=35,
        colorscale=[[0, 'darkblue'], [1, 'red']]
    )
)

fig2 = px.scatter_map(
    df_beauty,
    lat='latitude',
    lon='longitude',
    hover_name='city',
    color='beauty_score',
    size='sizes_score',
    color_continuous_scale=[[0, 'rgb(80,80,80)'], [1, 'rgb(255,255,0)']],
    zoom=4,
    map_style='carto-positron',
    hover_data={'latitude': False, 'longitude': False, 'sizes_score': False}
)

fig2.update_traces(
    customdata=df_beauty['hotels_text'],
    hovertemplate='<b>%{hovertext}</b><br>Beauty Score: %{marker.color:.1f}<br>%{customdata}<extra></extra>'
)

fig2.update_layout(
    title='Gradient de Beauty Score + Top 3 Hôtels',
    height=700,
    width=900,
    margin=dict(l=0, r=0, b=0, t=50)
)

fig1.show()
fig2.show()
```
